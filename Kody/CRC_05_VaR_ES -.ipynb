# Dystrybucje w statystyce

W statystyce i teorii prawdopodobiestwa **dystrybucja**  opisuje, jak prawdopodobiestwo rozkada si na rne moliwe wartoci zmiennej. 

Dystrybucje zazwyczaj przedstawia si na ukadzie z dwiema osiami. Pozioma (x) mwi o wartoci jakiej zmiennej np. wzrocie, dugoci, IQ czy stracie w portfelu. Za pionowa (y) moe mwi o czstotliwoci wystpowania w danym zbiorze, prawdopodobiestwie (czstotliwo/liczba wszystkich obserwacji - inaczej zwana relative frequency), czy gstoci.


 


**Dystrybucje s fundamentem zarzdzania ryzykiem**

## Rodzaje dystrybucji

### 1. Rozkad dyskretny (dystrybucja dyskretna)
- Zmienne przyjmuj skoczon lub przeliczalnie nieskoczon liczb wartoci.
- **Przykad:** rzut kostk  kademu oczku przypisane jest okrelone prawdopodobiestwo.
- Opisywany przez **funkcj masy prawdopodobiestwa** (PMF  *probability mass function*).

### 2. Rozkad cigy (dystrybucja ciga)
- Zmienne mog przyjmowa dowoln warto z przedziau liczbowego.
- **Przykad:** wzrost czowieka.
- Opisywany przez **funkcj gstoci prawdopodobiestwa** (PDF  *probability density function*).
- Prawdopodobiestwo okrela si jako pole pod krzyw PDF na danym przedziale.

## Przykadowe rozkady

### Rozkady dyskretne:
- Rozkad Bernoulliego  
- Rozkad dwumianowy  

- Rozkad Poissona

### Rozkady cige:
- Rozkad normalny 


- Rozkad wykadniczy
- Rozkad log-normalny

Kady rozkadw ma swoje parametry, swoje wasnoci i ksztaty. 

Dla dystrybucji normalnej to np. odchyelnie standardowe i rednia.

# Value at Risk i Expected Shortfall

# Value at Risk

**Value at Risk (VaR) - Warto Zagroona** to miara ryzyka, ktra okrela  **maksymaln oczekiwan strat** przy zadanym poziomie ufnoci i horyzoncie czasowym.
$$
\text{VaR}_\alpha = \inf \left\{ x \in \mathbb{R} : F(x) \geq \alpha \right\}
$$

gdzie:  
- $  \alpha\ $  poziom istotnoci (np. 0{,}05 dla 95% poziomu ufnoci),  
- $ \ F(x) \ $  dystrybuanta (CDF) rozkadu strat portfela,  
- $ \ x \ $  warto, dla ktrej prawdopodobiestwo straty nie przekracza $  \alpha\ $.

 Z prawdopodobiestwem $ \alpha \ $ strata nie przekroczy $ VaR_\alpha\ $ w danym okresie.

Np. Jeli VaR z alf 5% wynosi -1.64 mwi mi to: W 5 na 100 przypadkw strace wicej ni 1.64, albo w 95 przypadkach nie strace wicej ni 1.64

 **VaR wymaga zaoenia co do dystrybucji**

 Mog mie np: 
 1) Zaoenia o rozkadzie empirycznym (tzw. VaR nieparametryczny)
 2) Zaoenia o rozkadzie teoretycznym (tzw. VaR parametryczny)


 ### VaR rozkadu normalnego:

# Expected Shortfall
---


$$
\text{ES}_\alpha = \mathbb{E} \left[ L \mid L \geq \text{VaR}_\alpha \right]
$$

gdzie:  
- $  \text{ES}_\alpha  $  Expected Shortfall na poziomie istotnoci $ \ \alpha \ $,  
- $  \text{L}   $ - losowa zmienna (tu opisujca straty portfela),  
- $  \text{VaR}_\alpha  $  warto Value at Risk na poziomie $ \alpha \ $.


Expected Shortfall (ES), zwany te Conditional VaR (CVaR), to **rednia strata pod warunkiem, e strata przekroczy VaR**. ES pokazuje, jak due mog by **ekstremalne straty**  czyli te najgorsze 5% (lub inny poziom istotnoci).



### Porwnanie VaR vs ES



VaR mwi: *"Jakiej straty nie przekrocz z prawdopodobiestwem 95%?"*  
ES mwi: *"Jeli ju przekrocz VaR, to jak dua bdzie rednio strata?"*

### ES Parametryczny

#### Wzr na parametryczny Expected Shortfall (ES)  rozkad normalny

$$
\text{ES}_\alpha = -\mu - \sigma \cdot \frac{\phi(z_\alpha)}{\alpha}
$$

gdzie:  
- $ \mu $  rednia zwrotw,  
- $ \sigma $  odchylenie standardowe zwrotw,  
- $ \phi(z_\alpha) $  gsto standardowego rozkadu normalnego w punkcie $ z_\alpha $,  
- $ z_\alpha $  kwantyl standardowego rozkadu normalnego odpowiadajcy poziomowi istotnoci $ \alpha $,  
- $ \alpha $  poziom istotnoci.
